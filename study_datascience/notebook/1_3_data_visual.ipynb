## ライングラフ

```python
from matplotlib import pyplot as plt

years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.8]

# 折れ線グラフ
plt.plot(years, gdp,
         color='green',
         marker='o',
        linestyle='solid')
# タイトル
plt.title("normal GDP")

# Y軸にラベル追加
plt.ylabel("Billions of $")
plt.xlabel("years")
plt.show()
```

## bar plot

```python
movies = ["Aniie Hall","Ben-Hur", "Casablanca", "Gandhi", "West Side Story"]
num_oscars=[5,11,3,8,10]

# default 0.8のため、棒の幅を調整(0.1センタリング)
xs = [i + 0.1 for i, _ in enumerate(movies)]
print(xs)

## enumerateの解説
## enumerateは、要素と同時にインデックスを取得できる
## https://note.nkmk.me/python-for-enumerate-zip/


# X軸を[xs], 高さ[num_oscars]で棒グラフを作る
plt.bar(xs, num_oscars)

plt.ylabel("# of Academy Award")
plt.xlabel("# movie list")
plt.title("My Favorite Movies")

test= [i + 1 for i in range(5)]
print(test)

# X軸のラベルを映画名を中心に配置
plt.xticks([i + 0.1  for i, _ in enumerate(movies)], movies)

plt.show()
```
