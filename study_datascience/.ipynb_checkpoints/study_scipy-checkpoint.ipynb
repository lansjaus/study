```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy import ndimage
from scipy import misc
```

```python
face = misc.face(gray=True) #  sample画像をグレースケールで読み込み
face
```

```python
plt.imshow(face, cmap=plt.cm.gray)
```

## ぼかした画像作成

```python
blurred = ndimage.gaussian_filter(face.astype(np.float), sigma=3)
plt.imshow(blurred, cmap=plt.cm.gray);
```

## ぼかし画像にフィルタをかけてシャープネス

```python
filter_blurred = ndimage.gaussian_filter(blurred, sigma=1)
gain = 20
sharpned = blurred + gain * (blurred - filter_blurred)
plt.imshow(sharpned, cmap=plt.cm.gray)
```

### 返値を非表示するのには";"を最後に入力する
