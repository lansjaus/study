# 統計
統計とは、データを理解する数学的手法を指します。
統計を扱う分野は広大で深淵で、膨大です。さらっと表面をさらいます

## データの特徴を表す

```python
# 友達の数を非ストグラムで書く
import random
import math
import collections
from matplotlib import pyplot as plt

random.seed(1)
# 適当に乱数から友達の数を作成
ran_n = 1000
num_friends = [math.floor(random.random()*100) for i in range(ran_n)]
print(num_friends)

# ヒストグラム
friend_counts = collections.Counter(num_friends)
xs = range(1000)
ys = [friend_counts[x] for x in xs]
plt.bar(xs,ys)
plt.axis([-5,105,0,25])
plt.title("histgram of Friend Count")
plt.xlabel("# of firend count")
plt.ylabel("# of UU")
plt.show()
```

```python
random.seed(4)

# 適当に乱数から友達の数を作成
daily_minutes = [random.random() * 10 for i in range(ran_n)]
daily_minutes
```

## 代表値

```python
def mean(x):
    return sum(x) / len(x)

mean(num_friends)
```

```python
def median(v):
    n = len(v)
    sorted_v = sorted(v)
    midpoint = n // 2
    
    if n % 2 == 1:
        # 奇数のため、中央を返す
        return sorted_v[midpoint]
    else:
        # 偶数のため、中央の要素の平均
        lo = midpoint - 1
        hi = midpoint
        return (sorted_v[lo] + sorted_v[hi]) / 2
    
median(num_friends)
```

## 分位数

```python
def quantile(x,p):
    p_index = int(len(x)*p)
    return sorted(x)[p_index]

print(quantile(num_friends, 0.1))
print(quantile(num_friends, 0.2))
```

## 最頻値

```python
def mode(x):
    counts = collections.Counter(x)
    max_count = max(counts.values())
    if type(max_count) == int:
        return max_count
    else:
        return [x_i for x_i, count in counts.iteritems() if count == max_count]



mode(num_friends)
```

## 分散、範囲

```python
def data_range(x):
    return max(x) - min(x)

data_range(num_friends)
```

```python
def de_mean(x):
    x_bar = mean(x)
    return [x_i - x_bar for x_i in x]

def dot(v,w):
    return sum(v_i * w_i for v_i, w_i in zip(v,w)) 

def sum_of_squares(v):
    return dot(v,v)

def data_variance(x):
    n = len(x)
    deviations = de_mean(x)
    return sum_of_squares(deviations) / (n-1)

def standard_deviation(x):
    return math.sqrt(x)


result = data_variance(num_friends)
print(result)
print(standard_deviation(result))
```

## 相関

分散は、ある変数が平均からどの程度乖離しているかを計るもの。分散が大きい場合、値は広く散らばっています。
共分散は、２つの変数に対し拡張したもの。２つの変数がその平均からどの程度異なっているか測ります。ただし、値の大きさに影響されます。

```python
def covariance(x,y):
    n = len(x)
    # ２つのベクトルの内積をベクトルの要素数-1で除す
    return dot(de_mean(x), de_mean(y)) / n -1

covariance(num_friends, daily_minutes)
```
