# 3.1 matplotlib

## ライングラフ

```python
from matplotlib import pyplot as plt

years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.8]

# 折れ線グラフ
plt.plot(years, gdp,
         color='green',
         marker='o',
        linestyle='solid')
# タイトル
plt.title("normal GDP")

# Y軸にラベル追加
plt.ylabel("Billions of $")
plt.xlabel("years")
plt.show()
```

## bar plot

```python
movies = ["Aniie Hall","Ben-Hur", "Casablanca", "Gandhi", "West Side Story"]
num_oscars=[5,11,3,8,10]

# default 0.8のため、棒の幅を調整(0.1センタリング)
xs = [i + 0.1 for i, _ in enumerate(movies)]
print(xs)

## enumerateの解説
## enumerateは、要素と同時にインデックスを取得できる
## https://note.nkmk.me/python-for-enumerate-zip/


# X軸を[xs], 高さ[num_oscars]で棒グラフを作る
plt.bar(xs, num_oscars)

plt.ylabel("# of Academy Award")
plt.xlabel("# movie list")
plt.title("My Favorite Movies")

test= [i + 1 for i in range(5)]
print(test)

# X軸のラベルを映画名を中心に配置
plt.xticks([i + 0.1  for i, _ in enumerate(movies)], movies)

plt.show()
```

## ヒストグラム

注意

axisとグラフ位置の調整で利用する、値には注意。本来描画している状況と異なって描画してしまう可能性がある。
簡易なテストデータを作成し、意図した描画になるかテストするのが良い

```python
# 修正済
import collections
grades = [83, 95, 91, 87, 70, 0, 85, 28, 100, 67, 73, 77, 0, 11, 20, 12, 5]

# lambdaで簡易なbin関数
decile = lambda grade: grade // 10 * 10
histogram= collections.Counter(decile(grade) for grade in grades)

plt.bar([x + 5 for x in histogram.keys()], # 左に-4ずらす
       histogram.values(), # 値に合わせて自動調整
       8) # 棒グラフ幅=8

plt.axis([-5, 120,0,5])
plt.xticks([10 * i for i in range(12)]) # X軸ラベル10刻み : 10 ～ 110
plt.xlabel("Declie")
plt.ylabel("# of Students")
plt.title("Distibution of Exam 1 Grades")

plt.show()
```

## 折れ線グラフ

```python
variance = [1, 2, 4, 8, 16, 42, 64, 128, 256]
bias_squared = [256, 128, 64, 32, 16, 8, 4, 2, 1]
total_error = [x + y for x, y in zip(variance, bias_squared)]
xs = [i for i, _ in enumerate(variance)]

# pltを複数買い呼び出して複数の線を描画, label指定しているため凡例は自動
plt.plot(xs, variance, 'g', label='variance')
plt.plot(xs, bias_squared, 'r', label='bias^2')
plt.plot(xs, total_error, 'b', label='total error')

# 調整
plt.legend(loc=9)
plt.xlabel("model complexity")
plt.title("the Bias-Variance Tradeoff")
plt.show()
```

## 散布図

```python
friends = [70,   65,  72,  63,  71,  62,  72,  66, 65]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels = ['a','b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']
plt.scatter(friends, minutes)

for label, friend_count, minute_count in zip(labels, friends, minutes):
    plt.annotate(label,
                xy=(friend_count, minute_count),
                xytext=(5,-5),
                textcoords='offset points')

plt.title("Daily Minustes vs. Number of Frieneds")
plt.show()
```
