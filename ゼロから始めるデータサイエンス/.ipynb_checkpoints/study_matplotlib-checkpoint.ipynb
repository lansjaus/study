## 導入

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,100)
plt.plot(x, np.sin(x))
```

```python
x = np.linspace(0,10,100)
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

```python
plt.figure(figsize=(8,4), dpi=80) # サイズと解像度
x = np.linspace(0,10,100)
plt.plot(x, np.sin(x), color="red", linewidth=3, label="sin(x)")
plt.plot(x, np.cos(x), "b*", linewidth=3, label="cos(x)")
plt.xlim(0, np.pi * 2)
plt.ylim(-2,2)
plt.xticks([0, np.pi, 2 * np.pi], ["0", "$\pi$", "2$\pi$"])
plt.yticks([-1, 0, 1])
plt.grid()
plt.legend()
plt.title("sin / cos")
plt.xlabel("phase")
plt.savefig("sincos.png")
```

```python
x = np.linspace(0,10,100)
plt.subplot(211) # 2行1列名の1目
plt.plot(x, np.sin(x))
plt.subplot(212) # 2行1列名の2目
plt.plot(x, np.cos(x))
```
